```python
from unsloth import FastModel
import torch

model, tokenizer = FastModel.from_pretrained(
    model_name = "/content/drive/MyDrive/coder-next",
    max_seq_length = 1024,
    dtype = None,        # Keep compute in BF16
    load_in_4bit = False,          # Must be False for FP8
    load_in_fp8 = True,        # <--- ENABLE FP8 HERE
    local_files_only = True,
    trust_remote_code=True,
    device_map="balanced_low_0",

)
print("✅ Model loaded!")
```

```python
from unsloth import FastModel
from peft import get_peft_model, LoraConfig, TaskType

model = FastModel.get_peft_model(
    model,
    r = 128,                                      # Reduced for memory saving
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 128,                             # Reduced for memory saving
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
)

print("✅ LoRA adapter applied with r=64")
```

```python
from datasets import load_dataset

# Change the path and column name if needed
dataset = load_dataset("json", data_files="/content/tweakbot18k.jsonl", split="train")

print(f"✅ Dataset loaded: {len(dataset)} examples")
```

```python
import unsloth.tokenizer_utils
unsloth.tokenizer_utils.fix_untrained_tokens = lambda *args, **kwargs: None
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="messages",
    args=SFTConfig(
        output_dir="/content/tweakbot-checkpoints",
        per_device_train_batch_size=1,
        gradient_accumulation_steps=8,
        num_train_epochs=1,
        max_seq_length=1024,
        learning_rate=2e-4,
        optim="paged_adamw_8bit",
        gradient_checkpointing= "unsloth",
        save_strategy="steps",
        save_steps=200,
        logging_strategy="steps",
        logging_steps=10,
        warmup_steps=10,
        seed=3407,
    ),
)
print("✅ Trainer initialized,")
```

```python
trainer.train()
print("✅ Training Completed!")
```

```python
merged_model = model.merge_and_unload()
print("✅ LoRA merged into base model!")
```
